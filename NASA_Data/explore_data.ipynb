```python
from matplotlib import pyplot as plt
import pandas as pd
import os
```

```python
arinc_df = pd.read_parquet('translated/652200111111735_ARINC429_F0_translated/Engine_Fan_RPM_N1_ACTUAL_40.parquet/')
arinc_df['time'] = pd.to_datetime(arinc_df['time'])
arinc_df.set_index('time', inplace=True)
arinc_df
```

```python
for file in os.listdir('translated/652200111111735_ARINC429_F0_translated/'):
    if file.endswith('.parquet'):
        temp_df = pd.read_parquet(os.path.join('translated/652200111111735_ARINC429_F0_translated/', file))
        temp_df['time'] = pd.to_datetime(temp_df['time'])
        temp_df.set_index('time', inplace=True)
        arinc_df = arinc_df.merge(temp_df, on='time')
```
